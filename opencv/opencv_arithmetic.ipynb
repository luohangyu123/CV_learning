```python
import numpy as np
import cv2
# 以图的形式读取图像
from matplotlib import pyplot as plt
```

```python
rain = cv2.imread("../img/rain.jpg")
```

```python
view = cv2.imread("../img/view.jpg")
```

```python
plt.imshow(rain[:,:,::-1])
```

```python
plt.imshow(view[:,:,::-1])
```

```python
img1 = cv2.add(rain, view)
```

```python
plt.imshow(img1[:,:,::-1])
```

```python
img2 = rain + view
```

```python
plt.imshow(img2[:,:,::-1])
```

```python
img = np.full((400,400,3), 50, np.uint8)
```

```python
img3 = cv2.subtract(view, img)
```

```python
plt.imshow(img3[:,:,::-1])
```

```python
img4 = cv2.addWeighted(rain, 0.3, view, 0.7, 0)
```

```python
plt.imshow(img4[:,:,::-1])
```
