```python
import numpy as np
import cv2
# 以图的形式读取图像
from matplotlib import pyplot as plt
```

```python
dogsp = cv2.imread("../img/dogsp.jpeg")
```

```python
doggauss = cv2.imread("../img/dogGauss.jpeg")
```

```python
plt.imshow(dogsp[:,:,::-1])
```

```python
plt.imshow(doggauss[:,:,::-1])
```

```python
res1 = cv2.blur(dogsp, (10,10))
```

```python
plt.imshow(res1[:,:,::-1])
```

```python
res2 = cv2.blur(doggauss, (10,10))
```

```python
plt.imshow(res2[:,:,::-1])
```

```python
res3 = cv2.GaussianBlur(doggauss,(5,5),1)
```

```python
plt.imshow(res3[:,:,::-1])
```

```python
res4 = cv2.medianBlur(dogsp, 5)
```

```python
plt.imshow(res4[:,:,::-1])
```
