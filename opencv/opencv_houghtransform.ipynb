```python
import numpy as np
import cv2
# 以图的形式读取图像
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("../img/calender.jpg")
```

```python
# plt.imshow(img, cmap=plt.cm.gray)
```

```python
edges = cv2.Canny(img, 50, 100)
```

```python
plt.imshow(edges, cmap=plt.cm.gray)
```

```python
# r， theta的精确度，也就是步长
lines = cv2.HoughLines(edges, 0.8, np.pi/180, 180)
```

```python
lines
```

```python
for line in lines:
    rho, theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a * rho
    y0 = b * rho
    x1 = int(x0 + 1000 * (-b))
    y1 = int(y0 + 1000 * (a))
    x2 = int(x0 - 1000 * (-b))
    y2 = int(y0 - 1000 * (a))
    cv2.line(img, (x1, y1), (x2, y2), (0, 255, 0))
```
