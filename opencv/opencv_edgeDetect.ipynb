```python
import numpy as np
import cv2
# 以图的形式读取图像
from matplotlib import pyplot as plt
```

```python
horse = cv2.imread("../img/horse.jpg", 0)
```

```python
# Sobel_x_or_y = cv2.Sobel(src, ddepth, dx, dy, dst, ksize, scale, delta, borderType)
x = cv2.Sobel(horse, cv2.CV_16S, 1, 0)
y = cv2.Sobel(horse, cv2.CV_16S, 0, 1)
```

```python
Scale_absX = cv2.convertScaleAbs(x)  # convert 转换  scale 缩放
Scale_absY = cv2.convertScaleAbs(y)
```

```python
result = cv2.addWeighted(Scale_absX, 0.5, Scale_absY, 0.5, 0)
```

```python
plt.imshow(result, cmap=plt.cm.gray)
```

```python
x = cv2.Sobel(horse, cv2.CV_16S, 1, 0, ksize=-1)
y = cv2.Sobel(horse, cv2.CV_16S, 0, 1, ksize=-1)
```

```python
Scale_absX = cv2.convertScaleAbs(x)  # convert 转换  scale 缩放
Scale_absY = cv2.convertScaleAbs(y)
```

```python
result = cv2.addWeighted(Scale_absX, 0.5, Scale_absY, 0.5, 0)
```

```python
plt.imshow(result, cmap=plt.cm.gray)
```

```python
# 拉普拉斯算子
h = cv2.Laplacian(horse, cv2.CV_16S)
```

```python
res = cv2.convertScaleAbs(h)
```

```python
plt.imshow(res, cmap=plt.cm.gray)
```

```python
# canny算子
h2 = cv2.Canny(horse, 0, 100)
```

```python
res = cv2.convertScaleAbs(h2)
```

```python
plt.imshow(res, cmap=plt.cm.gray)
```
