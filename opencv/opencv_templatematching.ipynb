```python
import numpy as np
import cv2
# 以图的形式读取图像
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("../img/wulin.jpeg")
template = cv2.imread("../img/shu.png")
```

```python
plt.imshow(img[:,:,::-1])
```

```python
plt.imshow(template[:,:,::-1])
```

```python
# 模板匹配 res = cv.matchTemplate(img,template,method)
res = cv2.matchTemplate(img,template,method=cv2.TM_CCOEFF)
```

```python
plt.imshow(res, cmap=plt.cm.gray)
```

```python
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
```

```python
top_left = max_loc
```

```python
h, w = template.shape[:2]
low_right = (top_left[0]+w, top_left[1]+h)
```

```python
cv2.rectangle(img, top_left, low_right, (0,255,0), 2)
```

```python
plt.imshow(img[:,:,::-1])
```
