```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("../img/letter.png")
```

```python
plt.imshow(img[:,:,::-1])
```

## 1.腐蚀与膨胀

```python
kernel = np.ones((5, 5), np.uint8)
# 3 图像腐蚀和膨胀
erosion = cv2.erode(img, kernel) # 腐蚀
```

```python
kernel
```

```python
plt.imshow(erosion[:,:,::-1])
```

```python
dilate = cv2.dilate(img, kernel) # 膨胀
```

```python
plt.imshow(dilate[:,:,::-1])
```

## 2.开闭运算

```python
img1 = cv2.imread("../img/letteropen.png")
img2 = cv2.imread("../img/letterclose.png")
```

```python
kernel = np.ones((10, 10), np.uint8)
```

```python
open = cv2.morphologyEx(img1, cv2.MORPH_OPEN, kernel)
```

```python
plt.imshow(open[:,:,::-1])
```

```python
close = cv2.morphologyEx(img2, cv2.MORPH_CLOSE, kernel)
```

```python
plt.imshow(close[:,:,::-1])
```

## 3.礼帽与黑帽

```python
top_hat = cv2.morphologyEx(img1, cv2.MORPH_TOPHAT, kernel)
```

```python
plt.imshow(top_hat[:,:,::-1])
```

```python
black_hat = cv2.morphologyEx(img2, cv2.MORPH_BLACKHAT, kernel)
```

```python
plt.imshow(black_hat[:,:,::-1])
```
