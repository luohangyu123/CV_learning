```python
import numpy as np
import cv2
# 以图的形式读取图像
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("../img/cat.jpeg", 0)
```

```python
plt.imshow(img, cmap=plt.cm.gray)
```

```python
histr = cv2.calcHist([img],[0],None,[256],[0,256])
```

```python
plt.figure()
plt.plot(histr)
plt.show()
```

```python
mask = np.zeros(img.shape[:2], np.uint8)
```

```python
mask[400:650, 200:500] = 1
```

```python
mask
```

```python
plt.imshow(mask, cmap=plt.cm.gray)
```

```python
# 3.mask
masked_img = cv2.bitwise_and(img,img,mask = mask)
```

```python
plt.imshow(masked_img, cmap=plt.cm.gray)
```

```python
histr2 = cv2.calcHist([img],[0],mask,[256],[0,256])
```

```python
plt.figure()
plt.plot(histr2)
plt.show()
```

```python
# 直方图均衡化
dst = cv2.equalizeHist(img)
```

```python
plt.imshow(dst, cmap=plt.cm.gray)
```

```python
# 2. 创建一个自适应均衡化的对象，并应用于图像
cl = cv2.createCLAHE(2.0, (8,8))
clahe = cl.apply(img)
```

```python
plt.imshow(clahe, cmap=plt.cm.gray)
```
