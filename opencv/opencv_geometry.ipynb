```python
img = cv2.imread("../img/2.jpg")
```

```python
plt.imshow(img[:,:,::-1])
```

```python
res1 = cv2.resize(img, None, fx=0.5, fy=0.5)
```

## 2.图像平移

```python
matrix = np.float32([[1,0,100],[0,1,50]])
```

```python
res2 = cv2.warpAffine(img,matrix,(cols,rows))
```

```python
plt.imshow(res2[:,:,::-1])
```

## 3.图像旋转

```python
M = cv2.getRotationMatrix2D((cols/2, rows/2), 45, 0.5)
```

```python
res3 = cv2.warpAffine(img, M, (cols,rows))
```

```python
plt.imshow(res3[:,:,::-1])
```

## 4.仿射变换

```python
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[100,100],[200,50],[100,250]])
```

```python
M = cv2.getAﬃneTransform(pts1,pts2)
```

```python
M
```

```python
res4 = cv2.warpAffine(img, M, (cols, rows))
```

```python
plt.imshow(res4[:,:,::-1])
```

## 5.透射变换

```python
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[100,145],[300,100],[80,290],[310,300]])
```

```python
M = cv2.getPerspectiveTransform(pts1,pts2)
```

```python
M
```

```python
res5 = cv2.warpPerspective(img,M,(cols,rows))
```

```python
plt.imshow(res5[:,:,::-1])
```

## 6.图像金字塔

```python
up_img = cv2.pyrUp(img)
down_img = cv2.pyrDown(img)   
```

```python
cv2.imshow('enlarge', up_img)
cv2.imshow('original', img)
cv2.imshow('shrink', down_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
