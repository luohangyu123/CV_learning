```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../img/tv.jpg")
```

```python
orb = cv2.ORB_create(5000)
```

```python
kp = orb.detect(img,None)
```

```python
kp, des = orb.compute(img, kp)
```

```python
img2 = cv2.drawKeypoints(img,kp,None, flags=0)
```

```python
plt.imshow(img2[:,:,::-1])
```
