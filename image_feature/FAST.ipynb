```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../img/tv.jpg")
```

```python
fast = cv2.FastFeatureDetector_create(threshold=30)
```

```python
kp = fast.detect(img,None)
```

```python
img2 = cv2.drawKeypoints(img,kp,None,(0,0,255))
```

```python
plt.imshow(img2[:,:,::-1])
```
