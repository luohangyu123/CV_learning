```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../img/tv.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# 2 角点检测
corners = cv2.goodFeaturesToTrack(gray, 500, 0.01, 10)
```

```python
corners
```

```python
# 3 绘制角点
# cv2.circle()函数中的圆心坐标不能是浮点数的类型, 要转换成整数才行
for i in corners:
    x, y = i.ravel()
    cv2.circle(img, (int(x), int(y)), 2, (0, 0, 255), -1)
```

```python
plt.imshow(img[:,:,::-1])
```
